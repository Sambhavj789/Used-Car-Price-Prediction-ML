```python
import json
import os
import pandas as pd
```

## Reading all jsonl files

```python
data = []
files = os.listdir("./output")
for file in files:
    if ".jsonl" in file:
        with open(f"./output/{file}","r") as f:
            for lines in f:
                data.append(lines)
```

```python
len(data)
```

## Reading all json file data

```python
json_files = ["dehradun-data.json","gurgaon-data.json","lucknow-data.json"]
for file in json_files:
    with open(f"./output/{file}","r") as f:
        curr_data = json.load(f)
        data = data + curr_data
```

```python
len(data)
```

## Converting data to dataframe

```python
data2 = []
for i in data[:30931]:
    try:
        data2.append(eval(i))
    except:
        continue
```

```python
data2 = data2 + data[30932:]
```

```python
df = pd.DataFrame(data2)
```

```python
df.to_csv("./data/data.csv",index=None)
```
