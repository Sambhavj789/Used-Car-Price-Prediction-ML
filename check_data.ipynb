```python
import json
import os
```

```python
total = 0
total_unique = 0
for file in os.listdir("./scraping-data"):
    if ".json" in file:
        with open(f"./scraping-data/{file}","r") as f:
            data = json.load(f)
            print(f"{len(data)} records found in {file}")
            print(f"{len(set(data))} unique records found")
            print("="*50)
            total += len(data)
            total_unique += len(set(data))
print("="*50)
print(f"Total Cars Found: {total}")
print(f"Total Unique Cars: {total_unique}")
```
