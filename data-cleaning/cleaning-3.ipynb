```python
import numpy as np
import pandas as pd
import json
```

```python
df = pd.read_csv("../data/data_cleaned-v2.csv")
```

```python
df.shape
```

```python
spec_cols = []
for i in df.columns:
    if "spec_" in i:
        spec_cols.append(i)
```

```python
len(spec_cols)
```

```python
dublicate_cols = []
for col in spec_cols:
    if col.replace("spec_","") in df.columns:
        print(col)
        dublicate_cols.append(col.replace("spec_",""))
```

```python
len(dublicate_cols)
```

```python
df.drop(columns=dublicate_cols,inplace=True)
```

```python
df.shape
```

```python
na_values = df.isna().sum()
na_values[na_values > 25000].shape
```
