## Calculating weight for each of the safety features

```python
safety_features_df = df[safety_features + ["Price"]].copy()
```

```python
safety_features_df.corr()["Price"].sort_values(ascending=False)
```

```python
# python -m pip install xgboost
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.ensemble import RandomForestRegressor
```

```python
X = safety_features_df.drop(columns = "Price")
Y = safety_features_df["Price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, random_state=25
)
```

```python
model = XGBRegressor(n_estimators = 500,random_state = 25)
```

```python
model.fit(x_train,y_train)
```

```python
rf_model = RandomForestRegressor(n_estimators=500,random_state=25,max_depth=50)
```

```python
rf_model.fit(x_train,y_train)
```

```python
# For XGBoost
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
```

```python
# For Random Forest
y_train_pred_2 = rf_model.predict(x_train)
y_test_pred_2 = rf_model.predict(x_test)
```

```python
# For XGBoost
print("Traning: ",r2_score(y_train,y_train_pred))
print("Test: ",r2_score(y_test,y_test_pred))
```

```python
print("Traning: ",r2_score(y_train,y_train_pred_2))
print("Test: ",r2_score(y_test,y_test_pred_2))
```

```python
model.feature_importances_
```

```python
feature_imporatence = pd.Series(model.feature_importances_,index=X.columns)
```

```python
feature_imporatence.sort_values(ascending=False)
```

```python
safety_score = (X * feature_imporatence).sum(axis=1)
```

```python
safety_score.describe()
```

## Normalizing safety score

```python
min_safety_score = safety_score.min()
max_safety_score = safety_score.max()
safety_score = ((safety_score - min_safety_score) / (max_safety_score - min_safety_score)) * 10
```

```python
safety_score.describe()
```

```python
df["Safety_Score"] = safety_score
```
