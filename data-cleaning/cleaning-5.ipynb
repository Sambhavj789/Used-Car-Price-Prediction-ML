```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../data/data_cleaned-v4.csv")
```

```python
df.info()
```

```python
df[["overview_Fuel type","spec_Fuel Type"]].isna().sum()
```

```python
df.drop(columns=["spec_Fuel Type"],inplace=True)
```

```python
df["overview_Fuel type"].value_counts()
```

## Replacing all values into 5 broad categories

```python
# Petrol , Diesel, CNG, Electric, LPG
df["overview_Fuel type"] = df["overview_Fuel type"].replace("Petrol + CNG","CNG")
df["overview_Fuel type"] = df["overview_Fuel type"].replace("CNG + CNG","CNG")
df["overview_Fuel type"] = df["overview_Fuel type"].replace("Hybrid (Electric + Petrol)","Electric")
df["overview_Fuel type"] = df["overview_Fuel type"].replace("Mild Hybrid(Electric + Petrol)","Electric")
df["overview_Fuel type"] = df["overview_Fuel type"].replace("Mild Hybrid (Electric + Diesel)","Electric")
df["overview_Fuel type"] = df["overview_Fuel type"].replace("Petrol + LPG","LPG")
df["overview_Fuel type"] = df["overview_Fuel type"].replace("Diesel + CNG","CNG")
df["overview_Fuel type"] = df["overview_Fuel type"].replace("LPG + LPG","LPG")
df["overview_Fuel type"] = df["overview_Fuel type"].replace("Diesel + LPG","LPG")
df["overview_Fuel type"] = df["overview_Fuel type"].replace("Electric + Electric","Electric")
df["overview_Fuel type"] = df["overview_Fuel type"].replace("Plug-in Hybrid (Electric + Petrol)","Electric")
df["overview_Fuel type"] = df["overview_Fuel type"].replace("LPG + CNG","LPG")
df["overview_Fuel type"] = df["overview_Fuel type"].replace("Electric + CNG","Electric")
df["overview_Fuel type"] = df["overview_Fuel type"].replace("Electric + LPG","Electric")
df["overview_Fuel type"] = df["overview_Fuel type"].replace("CNG + LPG","CNG")
```

```python
df["overview_Fuel type"].value_counts() 
```

```python
df[["spec_Transmission","overview_Transmission"]].isna().sum()
```

```python
df["overview_Transmission"].value_counts()
```

```python
df.drop(columns=["spec_Transmission"],inplace=True)
```

## Cleaning overview_Transmission, Converting all values into Manual and Automatic

```python
def clean_transmission(val):
    val = str(val)
    try:
        return val.split(" ")[0]
    except:
        return val


df["Transmission"] = df["overview_Transmission"].apply(clean_transmission)
```

```python
df.drop(columns=["overview_Transmission"], inplace=True)
```

```python
df.Transmission = df.Transmission.replace("Automatic,","Automatic")
df.Transmission = df.Transmission.replace("Clutchless","Automatic")
```

```python
df.Transmission.value_counts()
```

```python
df.drop(columns=["overview_Color"], inplace=True)
```

```python
df["overview_Insurance"].value_counts()
```

## Cleaning overview_Insurance columns

```python
def clean_insurance(val):
    if val in ["Comprehensive","Third Party","Zero Depreciation"]:
        return "Yes"
    return "No"
df["Insurance"] =df.overview_Insurance.apply(clean_insurance)
```

```python
df.drop(columns=["overview_Insurance"], inplace=True)
```

```python
df["overview_Registration Type"].value_counts()
```

```python
def clean_registration_type(val):
    if val in ["Corporate", "Taxi", "Commercial Registration"]:
        return "Commercial"
    if val == "NotAvailable":
        return "Not Available"
    return val


df["Registration_Type"] = df["overview_Registration Type"].apply(
    clean_registration_type
)
```

```python
df.drop(columns=["overview_Registration Type"], inplace=True)
```

```python
df["overview_Car Available at"].value_counts()
```

```python
df["City"] = df["overview_Car Available at"].str.split(",").str[-1].str.strip()
```

```python
df.drop(columns=["overview_Car Available at"], inplace=True)
```

## Cleaning spec_Engine columns

```python
def clean_engine(val: str):
    try:
        return val.split(",")[0].split("cc")[0]
    except:
        return val
df["Engine(cc)"] = df["spec_Engine"].apply(clean_engine)
```

```python
df.drop(columns=["spec_Engine"],inplace = True)
```

```python
df["spec_Max Power (bhp@rpm)"].value_counts()
```

## Cleaning spec_Max Power (bhp@rpm) column, Extracting bhp from each of them

```python
def clean_max_power(val):
    val = str(val)
    if "@" in val:
        return val.split("@")[0].replace("bhp","").strip()
    else:
        return val.replace("bhp","").strip()
df["Max_Power(bhp)"] = df["spec_Max Power (bhp@rpm)"].apply(clean_max_power)
```

```python
df.drop(columns=["spec_Max Power (bhp@rpm)", "spec_Max Torque (Nm@rpm)"], inplace=True)
```

## Columns which still have some string chars

```python
df["Max_Power(bhp)"][
    ~df["Max_Power(bhp)"].str.isnumeric() & ~df["Max_Power(bhp)"].isna()
].value_counts()
```

```python
def clean_max_power_2(val):
    try:
        return val.split(" ")[0].strip()
    except:
        return val


df["Max_Power(bhp)"] = df["Max_Power(bhp)"].apply(clean_max_power_2)
```

```python
df["Max_Power(bhp)"][~df["Max_Power(bhp)"].str.isnumeric() & ~df["Max_Power(bhp)"].isna()].value_counts()
```

```python
df["spec_Mileage"].value_counts()
```

```python
def clean_mileage(val: str):
    try:
        val2 = val.split(":")[1]
        if "|" in val2:
            val2 = val2.split("|")[0]
        if "(City)" in val2:
            val2 = val2.split("(City)")[0]
        return val2.strip()
    except:
        return val
df["Mileage"] = df["spec_Mileage"].apply(clean_mileage)
```

```python
df.drop(columns=["spec_Mileage"], inplace=True)
```
