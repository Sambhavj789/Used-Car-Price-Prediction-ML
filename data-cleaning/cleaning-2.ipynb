```python
import numpy as np
import pandas as pd
import json
```

```python
df = pd.read_csv("../data/data_raw.csv")
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
cols_info = {}
for cols in df.columns:
    cols_info[cols] = {"null": str(df[cols].isna().sum()), "unique": str(df[cols].nunique())}
```

```python
with open("columns-info.json","w") as file:
    json.dump(cols_info,file)
```

```python
cols_info.items()
```

## Finding the columns which have only 1 unique value

```python
cols_with_single_value = []
for col in cols_info.keys():
    if int(cols_info[col]["unique"]) == 1:
        print(col)
        cols_with_single_value.append(col)
```

```python
len(cols_with_single_value)
```

## Dropping those columns

```python
df.drop(columns = cols_with_single_value,inplace=True)
```

```python
df.shape
```

## Removing all columns whose name has overview_ prefix as those columns already exist in dataset

```python
overview_cols = []
for cols in cols_info.keys():
    if "overview_" in cols:
        overview_cols.append(cols)
```

```python
overview_cols
```

```python
df[["Price","overview_Price"]].sample(5)
```

```python
df["overview_Price"].isna().sum()
```

```python
def get_price(row):
    if str(row["Price"]) == 'nan':
        return row["overview_Price"]
    return row["Price"]
df["Price"] = df.apply(get_price,axis=1)
```

```python
overview_cols
```

```python
remove_cols = ["Kilometer","Fuel Type","Registration year","Manufacturing Year","No. of owners","Transmission","Color","Car Available at","Insurance","Registration Type"]
df.drop(columns=remove_cols,inplace=True)
```
